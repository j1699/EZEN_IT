### 예제 문제
1. csv 풀더에 있는 sales records.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit' 컬럼 삭제
3. 컬럼의 이름을 변경 ['대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 인덱스를 초기화(기존의 인덱스는 삭제)

```python
import pandas as pd

df = pd.read_csv("../csv/Sales Records.csv")
df.head()
```

```python
df.drop(["Order ID", 'Total Revenue', 'Total Cost', 'Total Profit'], axis=1, inplace=True)
```

```python
df.columns = ['대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
df.sort_values(['대륙', '국가'], inplace=True)
```

```python
# 인덱스 초기화, 기존의 인덱스 삭제

# case 1
    # df.reset_index(drop=True)

# case 2
    # df.reset_index().drop(['index'], axis=1)

df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
# 총 단가, 총 원가, 총 이윤 컬럼을 생성

# 총단가 = 단가 * 판매단위
df['총 단가'] = df['단가'] * df['판매단위']
df.head()
```

```python
# 총 원가 = 원가 * 판매단위
# 총 이윤 = 총 단가 - 총 원가
df['총 원가'] = df['원가'] * df['판매단위']
df['총 이윤'] = df['총 단가'] - df['총 원가']
df.head()
```

# loc[] : 특정한 위치에 데이터 출력
    - 특정 행이나 특정 열을 추출할 때 사용
    - loc[행의 위치, 열의 위치]
# iloc[]  : 위치 기반으로 데이터 출력
    - 특정 행이나 특정 열을 추출할 때 사용
    - iloc[행의 위치 인덱스 값, 열의 위치 인덱스 값]

```python
df.drop([0], axis=0,inplace=True)
df.drop([2], axis=0,inplace=True)
```

```python
# 특정 행 추출
df.loc[0:2]
```

```python
df.iloc[0:2]
```

```python
# 특정 열 추출
df.loc[:,'국가':'단가']
```

```python
df.iloc[0:2, 1:9]
```

### 그룹화
- groupby(그룹화 기준이 되는 컬럼)

```python
df.groupby("국가").sum()[['총 이윤']]
```

```python
# 가장 많은 이윤을 내는 대륙은 어디인가?
test = df.groupby("대륙").sum()[['총 이윤']]
```

```python
test.sort_values('총 이윤', ascending=False, inplace=True)
```

```python
test.head(1)
```

```python
# 국가별 총 이윤이 높은 상위 10개의 국가
# 국가별 그룹화
test_1 = df.groupby('국가').sum()[['총 이윤']]
# 총 이윤을 기준으로 내림차순 정렬
group_df = test_1.sort_values('총 이윤', ascending=False)
# 상위 10개
top10 = group_df.head(10)
# 하위 10개
bottom10 = group_df.tail(10)
```
